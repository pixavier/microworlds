```python
import os
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"
import math
import tensorflow as tf
from tensorflow.keras import backend as K
import tensorflowjs as tfjs

xs = tf.constant([1, 2, 3, 4], dtype=tf.int32)
#ys = tf.constant([2, 5, 7, 8], dtype=tf.int32)
ys = tf.constant([[2, 5, 7, 8], [4, 3, 2, 1], [6, 3, 7, 8], [1, 0, 1, 2]], dtype=tf.int32)

# Definir el modelo

model = tf.keras.Sequential()


def custom_activation(x):
    return tf.keras.activations.tanh(x)
#    return K.switch(x >= 0, tf.math.multiply(x, 100), tf.math.multiply(x, 0))
#    return tf.keras.activations.mish(x)
#    return tf.keras.activations.softplus(x)


# Agregar capas al modelo
model.add(tf.keras.Input(shape=(1,)))
#model.add(tf.keras.layers.Dense(units=50, activation=custom_activation))
model.add(tf.keras.layers.Dense(units=200, activation="tanh"))
model.add(tf.keras.layers.Dense(units=200, activation="tanh"))
model.add(tf.keras.layers.Dense(units=4))

print(model.summary())

def custom_loss(val_true, val_pred):
    return tf.losses.mean_absolute_error(val_true, val_pred)
#    return tf.losses.mean_squared_error(val_true, val_pred)


optim = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optim, loss=custom_loss)

epochs = 2000

model.fit(xs, ys, epochs=epochs, batch_size=4, verbose=1)

# Guardar el modelo entrenado
model.save('model_min_nn_keras')
tfjs.converters.save_keras_model(model, 'model_min_nn')

```
